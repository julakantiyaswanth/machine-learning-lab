```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

data = pd.read_csv("/water_potability.csv")
X = data[['Turbidity']]
y = data['Potability']

class_0 = 0
class_1 = 1
class_0_data = data[data['Potability'] == class_0]
class_1_data = data[data['Potability'] == class_1]
selected_data = pd.concat([class_0_data, class_1_data])
X_selected = selected_data[['Turbidity']]
y_selected = selected_data['Potability']

X_train, X_test, y_train, y_test = train_test_split(X_selected, y_selected, test_size=0.2, random_state=0)

# Use a linear kernel for SVM
clf = SVC(kernel='linear')
clf.fit(X_train, y_train)

support_vectors = clf.support_vectors_
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)

predicted_classes = clf.predict(X_test)
print("Predicted Classes using clf.predict(X_test):", predicted_classes)

# Adjust the threshold as needed
your_custom_predictions = np.where(predicted_classes >= 0.5, 1, 0)
custom_accuracy = accuracy_score(y_test, your_custom_predictions)
print("Custom Accuracy using Logic:", custom_accuracy)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

data = pd.read_csv("/water_potability.csv")
X = data[['Turbidity']]
y = data['Potability']

# Convert 'Potability' column to numeric values
y_numeric = pd.to_numeric(y, errors='coerce')

class_0 = 0
class_1 = 1
class_0_data = data[data['Potability'] == class_0]
class_1_data = data[data['Potability'] == class_1]
selected_data = pd.concat([class_0_data, class_1_data])
X_selected = selected_data[['Turbidity']]
y_selected = selected_data['Potability']

# Convert y_selected to a 1D array
y_selected = y_selected.values.ravel()

X_train, X_test, y_train, y_test = train_test_split(X_selected, y_selected, test_size=0.2, random_state=0)

kernel_functions = ['linear', 'poly', 'rbf', 'sigmoid']
for kernel_function in kernel_functions:
    clf = SVC(kernel=kernel_function)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy with {kernel_function} kernel: {accuracy}")
```
